```python
import re
reg_price_detect = re.compile('([$₹£€]|INR|USD|GBP|EUR).{0,2}[+-]?[0-9]{1,3}(?:,?[0-9]{3})*(?:\.[0-9]{2})?|[+-]?[0-9]{1,3}(?:,?[0-9]{3})*(?:\.[0-9]{2})?.{0,2}([$₹£€]|INR|USD|GBP|EUR)')
reg_price_detect.search('this cost ₹ 25900'),reg_price_detect.search('this cost 25900')
```

```python
reg_price_amount = re.compile('[^0-9.]')
test='this cost ₹ 25900'
print [eval(reg_price_amount.sub('',test[match.start():match.end()])) for match in reg_price_detect.finditer(test)]
test='this cost  25900'
print [eval(reg_price_amount.sub('',test[match.start():match.end()])) for match in reg_price_detect.finditer(test)]
```
