```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
soup = BeautifulSoup(requests.get('https://www.ylang23.com/').text)
```

```python
lis = soup.find_all('li')
```

```python
lis_childs = [len(li.find_all('li')) for li in lis]
```

```python
lis_url = [li.find('a')['href'] if li.find('a') else '' for li in lis]
lis_url_text = [li.find('a').text if li.find('a') else '' for li in lis]
```

```python
lis_a_class = ['' if not li.find('a') else '' if not li.find('a').get('class') else ','.join(li.find('a').get('class') ) 
               for li in lis]
```

```python
tmp = pd.DataFrame.from_records({'li':lis,'childs':lis_childs,'url':lis_url,'url_text':lis_url_text,
                                 'a_class':lis_a_class})
```
